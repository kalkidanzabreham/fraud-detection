Load & Inspect

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

credit = pd.read_csv("../data/raw/creditcard.csv")
credit.info()
```

Class Distribution

```python
credit["Class"].value_counts(normalize=True) * 100
sns.countplot(x="Class", data=credit)
plt.title("Credit Card Fraud Distribution")
plt.show()
```

Amount vs Fraud

```python
sns.boxplot(x="Class", y="Amount", data=credit)
plt.yscale("log")
plt.title("Transaction Amount vs Fraud")
plt.show()
```
